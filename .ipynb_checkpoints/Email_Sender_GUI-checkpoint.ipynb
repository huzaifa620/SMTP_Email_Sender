```python
import PySimpleGUI as psg
import random, time, os, itertools, csv, openpyxl, smtplib
from tkinter import simpledialog
import tkinter as tk
from tkinter import ttk
from tkinter import *
from pathlib import Path
import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import Font, Color, Alignment, Border, Side
from openpyxl.worksheet.dimensions import ColumnDimension, DimensionHolder
from openpyxl.utils import get_column_letter
from tkinter import simpledialog
import PySimpleGUI as sg
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
from os.path import basename
from smtplib import SMTPAuthenticationError

theme_name_list = sg.theme_list()
layout = [[sg.Text("CHOOSE A FILE: ",font=('Ubuntu',12)), sg.Input(), sg.FileBrowse(key="-IN-",font=('Ubuntu',12))],[sg.Button("SUBMIT",font=('Ubuntu',12))]]
sg.theme(theme_name_list[random.randint(0, len(theme_name_list))])
###Building Window
window = sg.Window('My File Browser', layout, size=(600,150))
    
while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED or event=="Exit":
        window.close()
        break
        
    elif event == "SUBMIT":
        path=(values["-IN-"])
        window.close()
        break
header=[]
if '.xlsx' in path:
    xlsx_file = Path(path)
    wb_obj = openpyxl.load_workbook(xlsx_file)
    total_sheets=wb_obj.get_sheet_names()
    def callback():
        global texter
        texter=(cmb.get())
        window.destroy()
    
    if len(total_sheets)>1:
        window=Tk()
        window.title('Sheet Selection') 
        window.geometry('200x100')
        l1=Label(window,text="Select Sheet")
        l1.grid(column=0, row=0)
        cmb=ttk.Combobox(window,values=total_sheets,width=30)
        cmb.grid(column=0, row=1)
        cmb.current(0)
        btn=Button(window,text="OK",command=callback)
        btn.grid(column=0, row=2)

        window.mainloop()
        print(texter)
        sheet = wb_obj.get_sheet_by_name(texter)
    else:
        sheet = wb_obj.active
    
    for row in sheet.iter_rows(max_row=1):
        for cell in row:
            header.append(cell.value)   
        print()
    print(header)
if '.csv' in path:
    df = pd.read_csv(path)
    header=list(df.columns)
    print(header)
if len(header)==0:
    raise Exception("File does not have any column")
    

while True:
    sg.theme(theme_name_list[random.randint(0, len(theme_name_list))])
    #define layout
    layout=[[sg.Combo(header,default_value='Select Column',key='board',font=('Ubuntu',12),button_arrow_color ='black',size=(20,100))],
            [sg.Button('SELECT', font=('Ubuntu',12)),sg.Button('CANCEL', font=('Ubuntu',12))]]
    #Define Window
    win =sg.Window('Select the Column Of Emails.',layout,size=(400, 400),element_justification="center",text_justification="center",resizable=True)
    #Read  values entered by user
    e,v=win.read()
    con = False 
    print(e,v)
    if e == None or e == "CANCEL":
        print('ham')
        print("exit")
        win.close()
        con = True
        print(1)
    #     driver.close()
        break
    else:
        if v['board']=='Select Column':
            print('Select Column from the dropdown!')
            win.close()
            continue
        else:
            win.close()
            break
if con:
    pass
else:
    dir_path = os.path.dirname(os.path.realpath('__file__'))
    column_name=v['board']
    print(column_name)

    
if '.xlsx' in path:
    ColNames = {}
    Current  = 0
    for COL in sheet.iter_cols(1, sheet.max_column):
        ColNames[COL[0].value] = Current
        Current += 1
    total_emails=[]
    for row_cells in sheet.iter_rows(min_row=2, max_row=sheet.max_row):
        if (row_cells[ColNames[column_name]].value) != 'Not available':
            total_emails.append(row_cells[ColNames[column_name]].value)
    print(total_emails)

if '.csv' in path:
    total_emails=(df[column_name].tolist())
real_emails=[]
for i in total_emails:
    if '||' in i:
        for j in i.split('||'):
            real_emails.append(j)
    elif i=='not available':
        continue
    else:
        real_emails.append(i)

media=sg.popup_yes_no('DO YOU WANT TO ADD THE MEDIA FILES?')
print(media)
posts = []
if media=='Yes':
    sg.set_options(element_padding=(3, 3))
    total_size=0
    layouts =  [
        [sg.Text("BROWSE THE MEDIA FILES: ")],
        [sg.Input(key='-INPUT-'), sg.FilesBrowse()],
        [sg.Button('ADD'), sg.Button('SUBMIT')],
        [sg.Multiline(size=(80, 25), key='-MULTILINE-')],
    ]

    window = sg.Window("MEDIA FILES", layouts, finalize=True)
    multiline = window['-MULTILINE-']
    window['-INPUT-'].expand(expand_x=True)
    while True:
        event, values = window.read()
        if event in (sg.WIN_CLOSED, 'SUBMIT'):
            break
        elif event == 'ADD':
            path = values['-INPUT-']
            if(path==''):
                multiline.print(f'Please select any file')
                multiline.print(f'**********************')
            elif path not in posts:
                stats = os.stat(path)
                print('Size of file is', stats.st_size/(1024*1024), 'megabytes')
                total_size=total_size+stats.st_size/(1024*1024)
                add=True
                if total_size <=25:
                    posts.append(path)
                else:
                    add=False
                    total_size=total_size-stats.st_size/(1024*1024)
                    multiline.print('*********ALERT*************')
                    multiline.print('You file/files size exceeded 25MB')
                    multiline.print('You cant attach files more than size of 25 MB in an email')
                    multiline.print(f'********************************************************')
                if add==True:
                    multiline.print(f'Number of files selected: {len(posts)}')
                    multiline.print(f'File Added: {path}')
                    multiline.print(f'Total Size: {total_size}')
                    multiline.print(f'************************')

            else:
                multiline.print(f'You have already added the file')
                multiline.print(f'\n')
    window.close()
else:
    pass

layout = [
    [sg.Text(f'You are sending emails to {len(real_emails)} users.',font=('Ubuntu',12))],
    [sg.Text('EMAIL SUBJECT',font=('Ubuntu',12)), sg.InputText(key='subject',font=('Ubuntu',12))],
    [sg.Text('EMAIL ADDRESS     ',font=('Ubuntu',12)), sg.InputText(key='sender',font=('Ubuntu',12))],
    [sg.Text('Enter Password:', font=('Ubuntu', 12)), sg.InputText(key='password', password_char='*', font=('Ubuntu', 12))],
    [sg.Text('EMAIL CONTENT',font=('Ubuntu',12)),sg.Multiline(size=(50, 5), key='textbox',font=('Ubuntu',12))],
    [sg.Button('SEND',font=('Ubuntu',12)), sg.Button('EXIT',font=('Ubuntu',12))]
]

window1= sg.Window('Email Sender', layout).Finalize()

while True:
    event, values = window1.read()
    if event == sg.WIN_CLOSED or event == 'EXIT':
        window1.close()
    elif event == "SEND":
        email_text=(values['textbox'])
        email_subject=(values['subject'])
        email_sender_address=(values['sender'])
        password=(values['password'])
        window1.close()
        break

window1.close()
# print(real_emails)
# print(email_sender_address)
# print(email_subject)
# print(email_text)
# print(password)
# print(media)
# print(posts)

def send_mail(send_from: str, subject: str, text: str, send_to: list, files=None):
    global password
    send_to = default_address if not send_to else send_to
    msg = MIMEMultipart()
    msg['From'] = send_from
    msg['To'] = ', '.join(send_to)  
    msg['Subject'] = subject
    msg.attach(MIMEText(text))
    for f in files or []:
        with open(f, "rb") as fil: 
            ext = f.split('.')[-1:]
            attachedfile = MIMEApplication(fil.read(), _subtype = ext)
            attachedfile.add_header('content-disposition', 'attachment', filename=basename(f))
        msg.attach(attachedfile)

    if '@gmail' in send_from:
        smtp = smtplib.SMTP(host="smtp.gmail.com", port=587)
    elif '@yahoo' in send_from:
        smtp = smtplib.SMTP(host="smtp.mail.yahoo.com", port=587)
    elif '@outlook' in send_from or '@hotmail' in send_from:
        smtp = smtplib.SMTP(host="smtp-mail.outlook.com", port=587)
    elif '@verizonwireless' in send_from:
        smtp = smtplib.SMTP(host="smtp.verizon.net", port=465)

    valid_password = False
    smtp.starttls()
    while not valid_password:
        try:
            smtp.login(email_sender_address, password)
            valid_password = True
            smtp.sendmail(send_from, send_to, msg.as_string())
            smtp.close()
        except SMTPAuthenticationError:
            layout_pwd = [
                [sg.Text('Enter password again:', font=('Ubuntu', 12)), sg.InputText(key='password', password_char='*', font=('Ubuntu', 12))],
                [sg.Button('OK', font=('Ubuntu', 12)), sg.Button('CANCEL', font=('Ubuntu', 12))]
            ]
            window_pwd = sg.Window('Invalid Password', layout_pwd, modal=True)
            event, values = window_pwd.read()
            if event == sg.WIN_CLOSED or event == 'CANCEL':
                window_pwd.close()
                break
            else:
                password = values['password']
                window_pwd.close()
                continue

if email_sender_address and email_subject and email_text and real_emails and password:

    send_mail(send_from=email_sender_address,
              subject=email_subject,
              text=email_text,
              send_to=real_emails,
              files=posts
    )

    sg.popup('Email(s) sent successfully')
```
